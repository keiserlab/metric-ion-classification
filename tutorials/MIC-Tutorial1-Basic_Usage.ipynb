# Tutorial 1: Calculating fingerprints and making predictions with MIC 

***NOTE: By default, all fingerprints are prune-eifp fingerprints.*** 


There are two ways to calculate fingerprints with MIC. The first is to use the FPParser on a single PDB file to generate fingerprints for all residues in a file. 

````python
from io import StringIO
from mic.fingerprints.FPParser import FPParser
````

````python
fpparser = FPParser('4OKE.pdb', n_proc = 10)
````

````python
fpparser.get_as_dataframe()
````

````python
assert fpparser.errors == []
````

### Alternatively, fingerprints for individual stuctures can be generated within a single PDB or across multuple PDB by providing an "entries" file

Entries files require the following CSV-dilineated columns: PDB_ID,chain,residue_name,residue_number,skip_symmetry

```
4OKE,A,MG,202,
3A09,A,CA,601,
4L9P,B,ZN,601,
7aue,R,CA,501,X
7f53,R,CA,601,X
7f55,R,CA,601,X
```

````python
entry_strio = StringIO("""4OKE,A,MG,202,
3A09,A,CA,601,
4l9p,B,ZN,601,
7aue,R,CA,501,X
7F53,R,CA,601,X
7f55,R,CA,601,X""")
````

````python
fpparser = FPParser('', entry_file=entry_strio, n_proc = 10)
````

````python
fpparser.get_as_dataframe()
````

It is possible (but not recommended) to call process_entry explicitly. This returns a single numpy array corresponding to the requested entity, along with any raised errors and the cluster flag.

````python
from luna.mol.entry import Entry
from mic.fingerprints.luna_utils import process_entry
````

````python
entry = Entry('4oke.symexp', 'A', 'MG', 202)
````

````python
fp, errors, flag = process_entry(entry)
````

````python
fp, flag
````

# Predicting with MIC

The trained MIC models are available through the MIC object, which loads both the metric encoder and SVC. There are two available models: MIC-prevalent and MIC-extended.

````python
from mic.models.MIC import MIC
````

````python
mic_prevalent = MIC('prune-eifp')
mic_extended = MIC('prune-eifp', extended_labels=True)
````

````python
mic_prevalent.predict(fps.drop(4097, axis=1).values, entries = fps.index)
````
